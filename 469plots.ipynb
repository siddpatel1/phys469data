```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sns point plot for 24 hr
# sns stripplot for 7:00
```

```python
m1=pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/moon1.csv")
m1.head()
```

```python
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(9,5), dpi=175)
sns.stripplot(x="time", y="alt", data=m1, edgecolor="gray", alpha=0.7, palette="icefire_r", linewidth=0.5);
plt.title("Moon PLOT #2");
plt.xlabel("Time of Day");
plt.ylabel("Altitude in Degrees");
```

```python
plt.figure(figsize=(9,5), dpi=175)
sns.pointplot(x="time24", y="alt", data=m1, edgecolor="gray", alpha=0.7, palette="icefire_r", linewidth=0.5, ci="sd");
plt.title("Moon PLOT #1");
plt.xlabel("Time in hours: T_0 = 0 = 7:00 PM on 1/10/22");
plt.ylabel("Altitude in Degrees");
```
