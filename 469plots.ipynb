```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sns point plot for 24 hr
# sns stripplot for 7:00
```

```python
moon1 = pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/moon1.csv")
betelgeuse1 = pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/Betelgeuse1.csv")
castor1 = pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/Castor1.csv")
polaris1 = pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/Polaris1.csv")
GammaCassiopeiae1 = pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/GammaCassiopeiae1.csv")
Algenib1 = pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/Algenib1.csv")
Capella1 = pd.read_csv("https://raw.githubusercontent.com/siddpatel1/phys469data/main/Capella1.csv")
```

```python
def plot1(df, nm):
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(9,5), dpi=175, facecolor="white")
    sns.pointplot(x="time24", y="alt", data=df, alpha=0.7, 
              palette="icefire_r", ci="sd", capsize=0.3, errwidth=1.1);
    plt.title(nm+" PLOT #1");
    plt.xlabel("Time in hours: T_0 = 0 = 7:00 PM on 1/10/22");
    plt.ylabel("Altitude in Degrees");
```

```python
def plot2(df, nm):
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(9,5), dpi=175, facecolor="white")
    sns.stripplot(x="time", y="alt", data=df, alpha=0.7, 
              palette="icefire_r");
    plt.title(nm+" PLOT #2");
    plt.xlabel("Time of Day");
    plt.ylabel("Altitude in Degrees");
```

```python
objdfList = [moon1, betelgeuse1, castor1, polaris1, GammaCassiopeiae1, Algenib1, Capella1]
objnmList = ["Moon", "Betelgeuse", "Castor", "Polaris", "Gamma Cassiopeiae", "Algenib", "Capella"]
```

```python
for i in range(len(objdfList)):
    plot1(objdfList[i], objnmList[i])
    plot2(objdfList[i], objnmList[i])
```
